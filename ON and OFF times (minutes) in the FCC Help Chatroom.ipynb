# User chatting time (ON-chatting) in the FCC Help Room (Jan-Jul 2015)

Definitions:
* **OFF** from the chatroom if the user waits more than 5 minutes or more to send the next message
* **ON** if the user sent messages during the period before being considered off 
* assume the time for any first message to add 1 minute 

```python
%matplotlib inline
import os, sys
import matplotlib
import matplotlib.pyplot as plt
import pickle
import nltk
from datetime import datetime, date, timedelta
import re, math
from collections import Counter

#http://stackoverflow.com/questions/15173225/how-to-calculate-cosine-similarity-given-2-sentence-strings-python
WORD = re.compile(r'\w+')

def get_cosine(vec1, vec2): 
     intersection = set(vec1.keys()) & set(vec2.keys())
     numerator = sum([vec1[x] * vec2[x] for x in intersection])

     sum1 = sum([vec1[x]**2 for x in vec1.keys()])
     sum2 = sum([vec2[x]**2 for x in vec2.keys()])
     denominator = math.sqrt(sum1) * math.sqrt(sum2)

     if not denominator:
        return 0.0
     else:
        return float(numerator) / denominator

def text_to_vector(text):
     words = WORD.findall(text)
     #if len(words) == 0:
     #     words = ['EMPTY']
     return Counter(words)
```

After getting the data from disk, the following code (datausers dict):
* will visit all the messages per user
* it will make a score to evaluate if the current message is similar to itself or any other message after that
* if the following message for comparison is beyond the 5 minutes, it will stop the comparison and start evaluating the closest message

A internal iteration (usertimes dict) will take the score and evaluate if:
* it is the begining of an ON condition
* it is the last message before becoming OFF
* it is a solitary message

```python
directory = "/home/ec/Public/freecodecamp/analytics_project/data_analysis/"
raw = pickle.load(open(directory+"help.pkl", "rb"))
users = set()
for u in raw:
    if u['fromUser']['username'] not in users:
        users.add(u['fromUser']['username'])

# for u in users:
#     print(u)
#     break

# for elem in raw:
#     if elem['fromUser']['username'] == list(users)[5]:
#         username = elem['fromUser']['username']
#         break

print("USER DATA STARTED...")  
datausers = {}
usertimes = {}
#http://stackoverflow.com/questions/2788871/python-date-difference-in-minutes
fmt = '%Y-%m-%dT%H:%M'
for username in users:
     #CAPTURING TEXTS AND SCORES PER USER
     # initialization of username an data collector (messages list)
     if username not in list(datausers.keys()):
          datausers[username] = []
          #print(username)
     messages = []
     for i, elem in enumerate(raw):
         # loop over all messages and if it is from username, capture that text and time
         if elem['fromUser']['username'] == username:
             #print("OUTSIDE ",elem)
             text1 = elem['text']
             if WORD.search(text1) == None:
                text1 = 'BLANKorCHARACTERS'
             #texttime = datetime(int(elem['sent'][:4]),int(elem['sent'][5:7]),int(elem['sent'][8:10]),int(elem['sent'][11:13]),int(elem['sent'][14:16]))
             texttime = datetime.strptime(elem['sent'][:16],fmt)
             # prepare text for calculations and initialise comparisons
             vector1 = text_to_vector(text1)
             count = 0
             cosine = 0
             starttime = texttime
             for el in raw[i:]:
                 # have a look of this and following messages sent by username until there is a gap higher of 5 minutes between the current message and the last one
                 # (look if the person has been "on" for 5 minutes since this message)
                 if el['fromUser']['username'] == username:
                     #followingtime = datetime(int(el['sent'][:4]),int(el['sent'][5:7]),int(el['sent'][8:10]),int(el['sent'][11:13]), int(el['sent'][14:16]))
                     followingtime = datetime.strptime(el['sent'][:16],fmt)
                     if followingtime - starttime > timedelta(minutes=5):
                         #print(followingtime - starttime)
                         break
                     text2 = el['text']
                     if WORD.search(text2) == None:
                         text2 = 'BLANKorCHARACTERS'
                     #similarity scoring: cosine
                     vector2 = text_to_vector(text2)
                     cosine += get_cosine(vector1, vector2)
                     count += 1
             #print(count)
             # average all scores
             if count > 0: cosine = cosine/count
             text1_id = elem['id']
             # to append to messages list: the current text and id, the last text in the 5 minutes window, the time at which this text was sent, the score
             msg = (text1, text1_id, text2, elem['sent'][:16], cosine)
             messages.append(msg)
             
             #CAPTURING ON-OFF CHAT PERIODS
             if username not in list(usertimes.keys()):
                 usertimes[username] = []
             if len(usertimes[username]) == 0:
                 #first message sent: "on" message
                 usertimes[username].append([elem['sent'][:16]])
                 find_end = cosine
             # the person STILL in the chat room
             elif find_end <= .99:
                 if cosine > .99:
                      #if person "off" of chat room, attach to "on" message and set long activity to stop (find_end = 1)
                      usertimes[username][-1].append(elem['sent'][:16])
                      find_end = 1
             # it is a request to start chatting
             else:
                 # last first "on": attach to collection as first
                 usertimes[username].append([elem['sent'][:16]])
                 # if a long "on" activity, set to long activity: find_end = 0
                 if cosine <= .99:
                       find_end = 0
               
     
     #once completed to visit all messages of THAT person, attach to datausers     
     datausers[username] = messages
     #break
     
print("USER DATA COMPLETED...")    
```

The following section is to take the usertimes dict and count/calculate differences between beginning and end of ON individual periods when pertinent, to prepare for some descriptives

```python
onoff_chatdur = []
onoff_chats = 0
for username in users:
        registers = usertimes[username]
        onoff_chats = onoff_chats + len(registers)
        for t in registers:
                if t == 1:
                        onoff_chatdur.append(1)
                else:
                        timediff = datetime.strptime(t[-1],fmt) - datetime.strptime(t[0],fmt)
                        diff_minutes = timediff.seconds/60 + 1
                        if diff_minutes >= 1000: print(username, diff_minutes)
                        onoff_chatdur.append(diff_minutes)
total_min = sum(onoff_chatdur)/len(onoff_chatdur)
```
